### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np


fig, axs = plt.subplots(2, 2, figsize=(10, 8))


x_line = np.arange(0, 10)
y_line = x_line ** 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]


axs[0, 0].plot(x_line, y_line, marker='o', color='blue')
axs[0, 0].set_title('Line Plot: y = x^2')
axs[0, 0].set_xlabel('x values')
axs[0, 0].set_ylabel('y values')
axs[0, 0].grid(True)


axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Plot')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')


axs[1, 0].scatter(x_scatter, y_scatter, color='purple')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x values')
axs[1, 0].set_ylabel('y values')


axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=90, colors=['red', 'green', 'blue', 'orange'])
axs[1, 1].set_title('Pie Chart: Language Popularity')


plt.tight_layout()


plt.show()
```
